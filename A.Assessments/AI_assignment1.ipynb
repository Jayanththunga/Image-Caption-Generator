```python
import numpy as np
import pandas as pd
```

```python
# task1
data = np.random.rand(4,4)
df = pd.DataFrame(data, columns=['col1','col2','col3','col4'])
df
```

```python
# task2
# renaming df - columns:
new_column_names = {'col1': 'Random value 1',
                    'col2': 'Random value 2',
                    'col3': 'Random value 3',
                    'col4': 'Random value 4'}
df = df.rename(columns=new_column_names)
df
```

```python
# task3
# Find the descriptive statistics of the 'df' dataframe.
stats = df.describe()
stats
```

```python
# task4
# check for null values
null_values = df.isnull().sum()
print(null_values)

# Find data types
data_types = df.dtypes
print(data_types)
```

```python
# task 5
# Display the 'Random value 2' & 'Random value 3' columns with location method and index location method
columns_loc = df.loc[:, ['Random value 2', 'Random value 3']]
print(columns_loc)

# Display columns using integer-based index location (iloc)
columns_iloc = df.iloc[:, [1, 2]]
print(columns_iloc)
```
